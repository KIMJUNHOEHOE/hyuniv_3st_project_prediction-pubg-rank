### matchType 별로?

```python
IQR_list = ['assists', 'boosts', 'DBNOs', 'headshotKills', 'heals', 'kills',
       'killStreaks','revives','weaponsAcquired','longestKill','rideDistance','swimDistance', 'walkDistance']
IQR_dic = {}
for col in IQR_list :
    tmp_np = data[data[col]!=0 ][col]
    quartile_1, quartile_3 = np.percentile(tmp_np, [25, 75])
    iqr = quartile_3 - quartile_1
    upper_bound = quartile_3 + (iqr * 1.5)
    IQR_dic[col] = upper_bound
    del tmp_np
print(IQR_dic)
```

```python
zscore_list = ['assists', 'boosts', 'DBNOs', 'headshotKills', 'heals', 'kills',
       'killStreaks','revives','weaponsAcquired','longestKill','rideDistance','swimDistance', 'walkDistance']
zscore_dic = {}
for col in zscore_list:
    tmp_np = np.mean(data[col])+np.std(data[col])*4.
    zscore_dic[col] = tmp_np
print(zscore_dic)
```

```python
lst = ['assists', 'boosts', 'DBNOs', 'headshotKills', 'heals', 'kills',
       'killStreaks','revives','weaponsAcquired','longestKill','rideDistance','swimDistance', 'walkDistance']
max_dic ={}
for col in lst :
    if zscore_dic[col] >= IQR_dic[col]:
        max_dic[col] = zscore_dic[col]
    else :
        max_dic[col] = IQR_dic[col]
print(max_dic)
```

```python
# 이상치 제거 1
data_filter = data.copy()
for col in max_dic:
    data_filter = data_filter[data_filter[col] <= max_dic[col]]
```

```python
# 이상치 제거 2
data_filter.drop(index=data_filter[(data_filter['matchType']==1)&(data_filter['walkDistance']==0)&(data_filter['winPlacePerc']>=0.5)].index,inplace=True)
```

```python
data_filter.describe()
```

```python
# data 단위 변환
log_transform = ['longestKill','rideDistance','swimDistance', 'walkDistance']
for col in log_transform :
    data_filter[col+'_log'] = np.log(data_filter[col]+1)
```

```python
# data 변환 - headshotkills
data_filter['headshotKillrate'] = data_filter['headshotKills']/data_filter['kills']
data_filter['headshotKillrate'].fillna(value=0,inplace=True)
```

```python
# winPlacePerc 컬럼위치 변경
def change_column_order(df, col_name, index):
    cols = df.columns.tolist()
    cols.remove(col_name)
    cols.insert(index, col_name)
    return df[cols]
data_filter = change_column_order(data_filter,'winPlacePerc',len(list(data_filter.columns)))
```
